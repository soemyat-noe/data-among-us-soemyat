```python
# Column Descriptions - Before diving into the data, here’s a quick explanation of the column names used:

# Believe_Miracles: Do you believe in miracles? (Yes/No)
# Always_Romantic: Do you consider yourself romantic, even when things aren't going well? (Yes/No)
# Hearing_Colours: How much do you want to live in a world where we could feel colours? (Scale 1-5)
# Miyazaki_or_Lynch: Would you rather live in Hayao Miyazaki’s or David Lynch’s universe?
# True_Self: How much do you feel that your life reflects who you truly are? (Scale 1-5)
# Past_or_Future: Would you rather go back 400 years to the past or travel to the future?
# Abstract_or_Realistic: If your life were a painting, would it be more abstract or hyper-realistic?
# Popping_Bubbles: How much do you like the sound of bubble wrap popping? (Scale 1-5)
# Eating_Spicy: Do you keep eating spicy food even when it feels unbearable? (Yes/No)
# Moon_and_Sea: Do you think the moon has secret conversations with the sea? (Yes/No)
# Polish_Remover: How much do you like the smell of nail polish remover? (Scale 1-5)
#____________________________________________________________________________________________________________________________________________________
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'miracles_data.csv'
miracles = pd.read_csv(path)
```

```python
display(miracles)
```

```python
miracles.info()
```

```python
# As you can see above, I mainly asked Y/N questions with some 1-5 scale ones sprinkled in, for the sake of simplicity.
```

```python
# Summary statistics for numerical columns
miracles.describe()
```

```python
# The summary stats reveal that most people align on imaginative concepts like feeling colors, but sensory experiences like the smell of nail polish remover are more divisive.

# Brief summary of findings:
    # Hearing_Colours has a high average rating (4.33).
    # True_Self shows a median of 4, indicating alignment with self.
    # Popping_Bubbles has a wider range of opinions.
    # Polish_Remover has the most divided opinions.
```

```python
# Pie chart for Believe_Miracles
miracles['Believe_Miracles'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
plt.title('Do You Believe in Miracles?')
plt.show()
```

```python
# Bar chart for Hearing_Colours
miracles['Hearing_Colours'].value_counts().plot(kind='bar', color='lightblue')
plt.title('How Much Do You Want to Feel Colours?')
plt.ylabel('Amount of Votes')
plt.show()
```

```python
# Most people in the class believe in miracles and are excited by the concept of feeling colors, showing a strong alignment with abstract and imaginative ideas."
```

```python
# Bar chart for Polish_Remover
miracles['Polish_Remover'].value_counts().sort_index().plot(kind='bar', color='lightpink')
plt.title('Ratings for Polish Remover')
plt.ylabel('Amount of Votes')
plt.show()
```

```python
# Bar chart for Popping_Bubbles
miracles['Popping_Bubbles'].value_counts().sort_index().plot(kind='bar', color='lightblue')
plt.title('Ratings for Popping Bubbles')
plt.ylabel('Amount of Votes')
plt.show()
```

```python
# In contrast to imaginative concepts, sensory preferences are more divided. The smell of nail polish remover, in particular, evokes varied reactions, while bubble wrap popping is more generally liked.
```

```python
# Scatter plot for True_Self vs Hearing_Colours
miracles.plot.scatter(x='True_Self', y='Hearing_Colours', color='lightblue')
plt.title('Relationship Between True Self and Desire to Feel Colours')
plt.xlabel('True Self Rating')
plt.ylabel('Hearing Colours Rating')
plt.show()
```

```python
# Pie chart for Miyazaki_or_Lynch
miracles['Miyazaki_or_Lynch'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
plt.title('Would You Rather Live in Miyazaki or Lynch’s Universe?')
plt.show()
```

```python
# Bar chart for Abstract_or_Realistic
miracles['Abstract_or_Realistic'].value_counts().sort_index().plot(kind='bar', color='lightpink')
plt.title('If Your Life Were a Painting, Would It Be Abstract or Realistic?')
plt.xlabel('Rating')
plt.ylabel('Amount of Votes')
plt.show()
```

```python
# When it comes to self-reflection, most people feel somewhat connected to their true selves. Additionally, many express romantic tendencies even in hard times, and there is a balance between abstract and realistic views of life.
```

```python
# Correlation between Polish_Remover and Popping_Bubbles
correlation_sensory = miracles[['Polish_Remover', 'Popping_Bubbles']].corr()
display(correlation_sensory)
```

```python
# Scatter plot for Polish_Remover vs Popping_Bubbles
miracles.plot.scatter(x='Polish_Remover', y='Popping_Bubbles', color='lightblue')
plt.title('Relationship Between Polish Remover and Popping Bubbles Ratings')
plt.xlabel('Polish Remover Rating')
plt.ylabel('Popping Bubbles Rating')
plt.show()
```

```python
# Finally, we explore whether people who enjoy certain sensory experiences, like the smell of polish remover, also enjoy others, such as the sound of popping bubbles
```

```python
# Calculate the mean score for abstract concepts
abstract_mean = miracles[['Hearing_Colours']].mean()

# Calculate the mean score for sensory experiences
sensory_mean = miracles[['Polish_Remover', 'Popping_Bubbles']].mean().mean()

# Create a bar chart to compare the two
pd.Series([abstract_mean.values[0], sensory_mean], index=['Abstract', 'Sensory']).plot(kind='bar', color=['lightblue', 'lightpink'])
plt.title('Comparison of Abstract vs Sensory Preferences')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Overall, the class aligns strongly on abstract, imaginative concepts like believing in miracles and feeling colors. However, when it comes to sensory experiences, opinions are more divided, particularly around the smell of polish remover. This contrast highlights how subjective sensory experiences can be, compared to more universally appealing imaginative ideas.
```

```python
# Create a DataFrame with counts of each rating for True_Self, Abstract_or_Realistic, and Past_or_Future
stacked_data = pd.concat([miracles['True_Self'].value_counts().sort_index(),
                          miracles['Abstract_or_Realistic'].value_counts().sort_index(),
                          miracles['Past_or_Future'].value_counts().sort_index()], axis=1)

# Plot stacked bar chart
stacked_data.plot(kind='bar', stacked=True, color=['lightblue', 'lightgreen', 'lightpink'])
plt.title('Stacked Ratings for True Self, Abstract/Realistic, and Past/Future Preferences')
plt.ylabel('Count')
plt.show()
```

```python
# Now let's take a deeper look at how the class reflects on their lives, both in terms of how aligned they feel with their true selves and their broader worldview.
# We can compare how people view their lives in terms of abstract or realistic perspectives, and whether they prefer to look to the past or the future.
```

```python
# Create a DataFrame with counts of each rating for Miyazaki_or_Lynch, Hearing_Colours, and Polish_Remover
stacked_data = pd.concat([miracles['Miyazaki_or_Lynch'].value_counts().sort_index(),
                          miracles['Hearing_Colours'].value_counts().sort_index(),
                          miracles['Polish_Remover'].value_counts().sort_index()], axis=1)

# Plot stacked bar chart
stacked_data.plot(kind='bar', stacked=True, color=['lightblue', 'Silver', 'lightpink'])
plt.title('Stacked Ratings for Imaginative Worlds and Sensory Preferences')
plt.ylabel('Count')
plt.show()
```

```python
# This stacked bar chart reveals some interesting patterns. Most of the class feels moderately connected to their true selves, with a tendency to lean towards more realistic life perspectives.
# However, when it comes to time preferences, there’s a more diverse spread—some people would rather go back to the past, while others prefer to move forward into the future.
# This suggests that while people may be aligned with their self-perception, their outlook on time varies significantly."
```
