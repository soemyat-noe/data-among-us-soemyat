```python
# Column Descriptions - Before diving into the data, here’s a quick explanation of the column names used:

# Believe_Miracles: Do you believe in miracles? (Yes/No)
# Always_Romantic: Do you consider yourself a romantic, even when things aren't going well? (Yes/No)
# Hearing_Colours: How much do you want to live in a world where we could feel colours? (Scale 1-5)
# Miyazaki_or_Lynch: Would you rather live in Hayao Miyazaki’s or David Lynch’s universe?
# True_Self: How much do you feel that your life reflects who you truly are? (Scale 1-5)
# Past_or_Future: Would you rather go back 400 years to the past or to the future?
# Abstract_or_Realistic: If your life were a painting, would it be more abstract or hyper-realistic?
# Popping_Bubbles: How much do you like the sound of bubble wrap popping? (Scale 1-5)
# Eating_Spicy: Do you keep eating spicy food even when it feels unbearable? (Yes/No)
# Moon_and_Sea: Do you think the moon has secret conversations with the sea? (Yes/No)
# Polish_Remover: How much do you like the smell of nail polish remover? (Scale 1-5)
#____________________________________________________________________________________________________________________________________________________
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'miracles_data.csv'
miracles = pd.read_csv(path)
```

```python
display(miracles)
```

```python
miracles.info()
```

```python
# As you can see above, I mainly asked Y/N questions with a few rating from a scale of 1-5 sprinkled in to keep things straightforward. 
```

```python
miracles.describe()
```

```python
# The stats reveal that we mainly align on imaginative concepts like feeling colors, but sensory experiences like the smell of nail polish remover 
# are more divisive.

# Brief summary of findings:
    # Hearing_Colours has the highest average rating (4.33 out of 5).
    # True_Self shows a median of 4, which is a positive finding as this means most of us feel more alignment with our selves then not.
    # Popping_Bubbles has a wider range of opinions, with minium rating being and the max highest being 5 (out of 5).
    # Polish_Remover similarly has a wide range of divided opinions but the average rating is the lowest in this one.
```

```python
miracles['Believe_Miracles'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
plt.title('Do You Believe in Miracles?')
plt.ylabel('')
plt.show()
```

```python
# As I breifly mentioned earlier, most of the class alligns on our thoughts about abstract ideas like miracles, indicating a tendency towards
# hopefulness and imaginative thinking.

# This is further proven by other abstract concepts that we explore next.
```

```python
miracles['Always_Romantic'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightpink', 'lightblue'])
plt.title('Do You Consider Yourself Romantic, Even When Things Aren’t Going Well?')
plt.ylabel('')
plt.show()
```

```python
# The same percentage of the class also identifies as romantic, even during challenging times. I am interested in seeing if the same people who voted
# yes for miracles also voted yes for being a romantic.
```

```python
miracles['Hearing_Colours'].value_counts().plot(kind='bar', color='lightblue')
plt.title('How Much Do You Want to Feel Colours?')
plt.ylabel('Count of Responses')
plt.show()
```

```python
# The idea of feeling colours resonates strongly with most of the team, even further highlighting the class's inclination towards imaginative 
# and abstract concepts.

# We will explore the data points of the opinions towards divisive sensory experiences.
```

```python
miracles['Polish_Remover'].value_counts().plot(kind='bar', color='lightpink')
plt.title('Ratings for Polish Remover')
plt.ylabel('Count of Responses')
plt.show()
```

```python
# Responses to the smell of polish remover are the most divided, which does make sense as sensory experiences are more more subjective compared 
# to abstract concepts, especially something as a specific as the smell of polish remover.
```

```python
miracles['Popping_Bubbles'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Ratings for Popping Bubbles')
plt.ylabel('Count of Responses')
plt.show()
```

```python
# The sound of popping bubble wrap, on the other hand, recieved more consistent enjoyment, suggesting that certain sensory experiences are
# universally pleasing.
```

```python
stacked_data = pd.concat([miracles['True_Self'].value_counts().sort_index(),
                          miracles['Abstract_or_Realistic'].value_counts().sort_index(),
                          miracles['Past_or_Future'].value_counts().sort_index()], axis=1)

stacked_data.plot(kind='bar', stacked=True, color=['lightblue', 'Silver', 'lightpink'])
plt.title('Stacked Ratings for True Self, Abstract/Realistic, and Past/Future Preferences')
plt.ylabel('Count of Responses')
plt.show()
```

```python
miracles['Eating_Spicy'].value_counts().plot(kind='bar', color='lightpink')
plt.title('Would You Continue Eating Spicy Food Even When It’s Unbearable?')
plt.ylabel('Count of Responses')
plt.xticks(rotation = 0)
plt.show()
```

```python
# Interestingly, most of the team show resilience when it comes to spicy food, suggesting a willingness to push sensory limits for enjoyment and thrill.
```

```python
miracles['True_Self'].value_counts().plot(kind='bar', color='lightblue')
plt.title('How Much Do You Feel That Your Life Reflects Your True Self?')
plt.xticks(rotation = 0)
plt.ylabel('Count of Responses')
plt.show()
```

```python
# Most of us feel moderately connected to our true selves, though there’s still some variation, showing that self-reflection can be complex. 
# This is the only question that I didn't recieve a 1 or a 5 as an answer. 
```

```python
miracles['Abstract_or_Realistic'].value_counts().plot(kind='bar', color='lightpink')
plt.title('Would Your Life Be More Abstract or Hyper-Realistic?')
plt.ylabel('Count of Responses')
plt.xticks(rotation = 0)
plt.show()
```

```python
# Unsurprisingly, the class leans more towards abstraction!
```

```python
stacked_data = pd.concat([miracles['True_Self'].value_counts().sort_index(),
                          miracles['Abstract_or_Realistic'].value_counts().sort_index(),
                          miracles['Past_or_Future'].value_counts().sort_index()], axis=1)

stacked_data.plot(kind='bar', stacked=True, color=['lightblue', 'silver', 'lightpink'])
plt.title('Stacked Ratings for True Self, Abstract/Realistic, and Past/Future Preferences')
plt.ylabel('Count of Responses')
plt.show()
```

```python
# Most of the team seem to be future-focused, suggesting they might believe that their ideal version of themselves or the world will exist 
# at a later time, rather than now.
```

```python
miracles['Miyazaki_or_Lynch'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
plt.title('Would You Rather Live in Miyazaki or Lynch’s Universe?')
plt.ylabel('')
plt.show()
```

```python
# Bonus -
# Most respondents prefer to live in the whimsical, imaginative world of Hayao Miyazaki over the surreal, darker world of David Lynch, reinforcing the class’s preference for lighthearted, creative thinking.
```

```python
# In conclusion, self-reflection reveals that the team as a whole generally leans towards abstract views, while sensory and 
# reflective experiences differ widely. Fun, imaginative, creative thinkers!
```
